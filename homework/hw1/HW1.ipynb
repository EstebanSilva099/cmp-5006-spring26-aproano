Su simplicidad y su solidez criptográfica, comparada con la sustitución simple y el Vigènere (un cifrado de sustitución polialfabética), lo convirtieron en un éxito inmediato como cifrado de campo, utilizado por los británicos en la Guerra de los Bóers y la Primera Guerra Mundial, y por varias fuerzas armadas como cifrado de emergencia durante la Segunda Guerra Mundial.

## Que es

La máquina Enigma fue un dispositivo de cifrado utilizado por el ejército alemán durante la Segunda Guerra Mundial, desarrollado originalmente por el ingeniero Arthur Scherbius en 1918 para la seguridad de las comunicaciones comerciales. Esta máquina permitía a los operadores codificar mensajes mediante una serie de ruedas giratorias, y el receptor necesitaba conocer la configuración específica para decodificar la comunicación. Su seguridad se basaba en la combinación de rotores, reflector y plugboard, que generaban una enorme cantidad de configuraciones posibles.

## Lógica del rotor

Los rotores eran discos internos con cableado eléctrico que realizaban sustituciones de letras.

**Componentes de el rotor:**


*  26 contactos de entrada (uno por cada letra).
*  26 contactos de salida.
*  Un cableado interno que conectaba las letras en un orden específico.

Al presionar una tecla, la señal electrica entraba por cada uno de los rotores cada uno de los cuales realiza una sustitución y luego pasaba a el reflector que codifica aún más lla señal antes de regresar por los rotores en sentido contrariopara iluminar la salida cifrada en el tablero de luz. Permite que la señal vuelva a los rotores, garantizando que el cifrado sea reversible.

**Funcionamiento:**

Cada vez que se tocaba una tecla:

*   El primer rotor giraba una posición.
*   Como un odómetro, cuando completaba una vuelta, hacía avanzar el siguiente rotor.
*   Esto cambiaba constantemente el sistema de sustitución.

la misma letra nunca se cifraba igual dos veces consecutivas.

Enigma era un sistema polialfabético dinámico, mucho más complejo que Vigenère





## El reflector

**Funcionalidad:**



*   Recibía la señal después de pasar por los rotores.
*   ELa devolvía por un camino diferente a través de los mismos rotores.

la corriente eléctrica hacía un recorrido de ida y vuelta que codifica aún más la señal antes de regresar.

Una letra nunca podía cifrarse como sí misma. El reflector conectaba pares de letras (la letra obtenida después de pasar por todos los rotores a su equivalencia en las conexiones de redireccionamiento).

Si la señal llegaba a una letra (letra de el mensaje que se quiere cifrar), el reflector la enviaba obligatoriamente a otra diferente. Luego al regresar por los rotores, el camino era distinto al original.

Ninguna letra pudiera transformarse en sí misma

## tablero de conexiones

El tablero de conexiones, o Steckerbrett, es un mecanismo de permutación adicional presente en algunos modelos de Enigma, que permite a los operadores intercambiar pares de letras conectando cables en diferentes posiciones, aumentando así el número de configuraciones posibles y ampliando el espacio de claves. Cada cable intercambiaba dos letras antes y después de pasar por los rotores.

La configuración inicial de los rotores, su orden, la posición de los anillos alfabéticos y las conexiones del tablero de conexiones deben ser idénticas en ambas máquinas comunicantes para garantizar un cifrado y descifrado exitosos.

3. Euler's Totient Function

```python
def phi(n):
    result = n
    p = 2
    temp = n

    while p * p <= temp:
        if temp % p == 0:
            while temp % p == 0:
                temp //= p
            result -= result // p
        p += 1

    # If temp > 1, then temp is a prime factor > sqrt(n)
    if temp > 1:
        result -= result // temp

    return result

print(f"a) phi(21) = {phi(21)}")
print(f"b) phi(17) = {phi(17)}")
print(f"c) phi(360) = {phi(360)}\n")
```

Infiriendo la key

```python
def is_alpha(ch):
    o = ord(ch)
    return (65 <= o <= 90) or (97 <= o <= 122)

def to_upper(ch):
    o = ord(ch)
    if 97 <= o <= 122:  # a-z
        return chr(o - 32)
    return ch

def letter_to_num_A0(ch_upper):
    return ord(ch_upper) - 65

def num_to_letter_A0(n):
    return chr(n + 65)

def derive_keystream(ciphertext, known_plaintext):
    ks = []
    i = 0
    j = 0

    while i < len(ciphertext) and j < len(known_plaintext):
        c = ciphertext[i]
        p = known_plaintext[j]

        if is_alpha(c) and is_alpha(p):
            cU = to_upper(c)
            pU = to_upper(p)
            cN = letter_to_num_A0(cU)
            pN = letter_to_num_A0(pU)
            kN = (cN - pN) % 26
            ks.append(num_to_letter_A0(kN))
            i += 1
            j += 1
        elif (not is_alpha(c)) and (not is_alpha(p)):
            i += 1
            j += 1
        elif (not is_alpha(c)) and is_alpha(p):
            i += 1
        else:
            j += 1

    if len(ks) == 0:
        raise Exception("No se pudo derivar keystream")
    return ''.join(ks)

def smallest_repeating_key(keystream):
    n = len(keystream)
    for L in range(1, n + 1):
        ok = True
        for i in range(n):
            if keystream[i] != keystream[i % L]:
                ok = False
                break
        if ok:
            return keystream[:L]
    return keystream

def vigenere_decrypt(ciphertext, key):
    keyU = ""
    for ch in key:
        if is_alpha(ch):
            keyU += to_upper(ch)

    if len(keyU) == 0:
        raise Exception("La clave no puede ser vacia.")

    out = []
    key_pos = 0
    key_len = len(keyU)

    for ch in ciphertext:
        if is_alpha(ch):
            base = 65 if (65 <= ord(ch) <= 90) else 97
            cN = ord(ch) - base
            kN = letter_to_num_A0(keyU[key_pos % key_len])
            pN = (cN - kN) % 26
            out.append(chr(pN + base))
            key_pos += 1
        else:
            out.append(ch)

    return ''.join(out)

if __name__ == "__main__":
    ciphertext = "rgnoDVD{O0NU_WQ3_G1G3O3T3_A1AH3S_cc82272b}"
    known_plaintext = "picoCTF{"

    keystream = derive_keystream(ciphertext, known_plaintext)
    print("Keystream (del prefijo):", keystream)

    key = smallest_repeating_key(keystream)
    print("Clave minima inferida:", key)

    plaintext = vigenere_decrypt(ciphertext, key)
    print("Descifrado:", plaintext)
```

## GuessMyCheese

```python
import hashlib

TARGET_HASH = "b2fec14855045065cb3fdc056036270e7af50c46662f34acddb87e98a834a6af"
CHEESE_FILE = "cheese_list.txt"

def sha256_hex(b: bytes) -> str:
    return hashlib.sha256(b).hexdigest()

def load_cheeses(path: str) -> list[str]:
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        return [line.strip() for line in f if line.strip()]

def main():
    cheeses = load_cheeses(CHEESE_FILE)

    encodings = ["utf-8", "latin-1", "utf-16-le", "utf-16-be"]
    case_variants = [
        ("original", lambda s: s),
        ("lower", lambda s: s.lower()),
        ("upper", lambda s: s.upper()),
    ]

    for cheese in cheeses:
        for case_name, case_fn in case_variants:
            cv = case_fn(cheese)
            for enc in encodings:
                try:
                    cheese_bytes = cv.encode(enc)
                except Exception:
                    continue

                for salt in range(256):
                    salt_raw = bytes([salt])
                    salt_hex = f"{salt:02x}".encode(enc, errors="ignore")

                    candidates = [
                        ("append_raw", cheese_bytes + salt_raw),
                        ("prepend_raw", salt_raw + cheese_bytes),
                        ("append_hex", cheese_bytes + salt_hex),
                        ("prepend_hex", salt_hex + cheese_bytes),
                    ]

                    for method, data in candidates:
                        if sha256_hex(data) == TARGET_HASH:
                            print("MATCH!")
                            print("Cheese:", cheese)
                            print("Case:", case_name, "| Encoding:", enc)
                            print(f"Salt: {salt} (hex: {salt:02x})")
                            print("Method:", method)
                            return

    print("No match.")

if __name__ == "__main__":
    main()
```

```python
import json
import math
import random
import time
import re

# =========================
# CONFIG
# =========================
CIPHERTEXT = r"""
nafyffoxenefufytpqnafymfppfentkpxeafbaxraezaqqpzqgswnfyefzwyxnhzqgsfnxnxqlexlzpwbxlrzhkfystnyxqntlbwezhkfyzatppflrfnafefzqgsfnxnxqlevqzwesyxgtyxphqlehenfgetbgxlxenytnxqlvwlbtgflntpemaxzatyfufyhwefvwptlbgtycfntkpfecxppeaqmfufymfkfpxfufnafsyqsfyswysqefqvtaxraezaqqpzqgswnfyefzwyxnhzqgsfnxnxqlxelqnqlphnqnftzautpwtkpfecxppekwntpeqnqrfnenwbflnexlnfyfenfbxltlbfozxnfbtkqwnzqgswnfyezxflzfbfvflexufzqgsfnxnxqletyfqvnflptkqyxqwetvvtxyetlbzqgfbqmlnqywllxlrzafzcpxenetlbfofzwnxlrzqlvxrezyxsneqvvflefqlnafqnafyatlbxeaftuxphvqzwefbqlfospqytnxqltlbxgsyquxetnxqltlbqvnflatefpfgflneqvspthmfkfpxfuftzqgsfnxnxqlnqwzaxlrqlnafqvvflexuffpfgflneqvzqgswnfyefzwyxnhxenafyfvqyftkfnnfyufaxzpfvqynfzafutlrfpxegnqenwbflnexltgfyxztlaxraezaqqpevwynafymfkfpxfufnatntlwlbfyentlbxlrqvqvvflexufnfzalxiwfexefeeflnxtpvqygqwlnxlrtlfvvfznxufbfvfleftlbnatnnafnqqpetlbzqlvxrwytnxqlvqzweflzqwlnfyfbxlbfvflexufzqgsfnxnxqlebqfelqnpftbenwbflnenqclqmnafxyflfghtefvvfznxufphtenftzaxlrnafgnqtznxufphnaxlcpxcftltnntzcfysxzqznvxetlqvvflexufphqyxflnfbaxraezaqqpzqgswnfyefzwyxnhzqgsfnxnxqlnatneffcenqrflfytnfxlnfyfenxlzqgswnfyezxflzftgqlraxraezaqqpfyenftzaxlrnafgflqwratkqwnzqgswnfyefzwyxnhnqsxiwfnafxyzwyxqexnhgqnxutnxlrnafgnqfospqyfqlnafxyqmltlbfltkpxlrnafgnqkfnnfybfvflbnafxygtzaxlfenafvptrxesxzqZNV{L6Y4G_4L41H515_15_73B10W5_8F1KV808}
""".strip()

QUADGRAM_JSON_PATH = "EN.json"
ALPHABET = "abcdefghijklmnopqrstuvwxyz"

MAX_ROUNDS = 4000
CONSOLIDATE = 3
SEED = 123
STOP_ON_FLAG = True

# =========================
# UTILIDADES
# =========================
def load_quadgrams(path):
    with open(path, "r", encoding="utf-8") as f:
        obj = json.load(f)
    return obj["alphabet"], obj["quadgrams"]

def build_char_to_index(alphabet):
    return {ch: i for i, ch in enumerate(alphabet)}

def text_to_indices(text, alphabet_map):
    out = []
    for ch in text.lower():
        if ch in alphabet_map:
            out.append(alphabet_map[ch])
    return out

def decode_with_key(ciphertext, key, alphabet_map, alphabet):
    inv = [0] * len(alphabet)
    for plain_idx, cipher_idx in enumerate(key):
        inv[cipher_idx] = plain_idx

    res = []
    for ch in ciphertext:
        lo = ch.lower()
        if lo in alphabet_map:
            c_idx = alphabet_map[lo]
            p_idx = inv[c_idx]
            p_ch = alphabet[p_idx]
            res.append(p_ch.upper() if ch.isupper() else p_ch)
        else:
            res.append(ch)
    return "".join(res)

def fitness_from_plain_indices(plain_idxs, quadgrams):

    if len(plain_idxs) < 4:
        return -1e18

    s = 0
    quad_idx = (plain_idxs[0] << 10) + (plain_idxs[1] << 5) + plain_idxs[2]
    for x in plain_idxs[3:]:
        quad_idx = ((quad_idx & 0x7FFF) << 5) + x
        s += quadgrams[quad_idx]
    return s

def hill_climb(key, cipher_idxs, positions_by_cipher_char, quadgrams):

    inv = [0] * len(key)
    for p, c in enumerate(key):
        inv[c] = p
    plain = [inv[c] for c in cipher_idxs]

    best = fitness_from_plain_indices(plain, quadgrams)
    keys_tried = 0

    improved = True
    n = len(key)

    while improved:
        improved = False
        for p1 in range(n - 1):
            for p2 in range(p1 + 1, n):
                c1 = key[p1]
                c2 = key[p2]

                for idx in positions_by_cipher_char[c1]:
                    plain[idx] = p2
                for idx in positions_by_cipher_char[c2]:
                    plain[idx] = p1

                keys_tried += 1
                score = fitness_from_plain_indices(plain, quadgrams)

                if score > best:
                    best = score
                    improved = True
                    key[p1], key[p2] = key[p2], key[p1]
                else:
                    for idx in positions_by_cipher_char[c1]:
                        plain[idx] = p1
                    for idx in positions_by_cipher_char[c2]:
                        plain[idx] = p2

    return best, keys_tried, key

# =========================
# MAIN
# =========================
def break_substitution(ciphertext, quadgram_json_path, max_rounds=4000, consolidate=3, seed=123):
    random.seed(seed)

    q_alphabet, quadgrams = load_quadgrams(quadgram_json_path)
    alphabet = q_alphabet
    if len(alphabet) > 32:
        raise ValueError("El alfabeto del JSON debe ser <= 32 para el packing 5-bit.")

    amap = build_char_to_index(alphabet)
    cipher_idxs = text_to_indices(ciphertext, amap)
    if len(cipher_idxs) < 4:
        raise ValueError("Ciphertext demasiado corto (necesito >= 4 letras).")

    positions = [[] for _ in range(len(alphabet))]
    for i, c in enumerate(cipher_idxs):
        positions[c].append(i)

    start = time.time()
    n = len(alphabet)

    best_key = list(range(n))
    best_score = -1e18
    best_hit = 0
    total_keys = 0
    rounds_done = 0

    for r in range(max_rounds):
        rounds_done = r + 1
        key = list(range(n))
        random.shuffle(key)

        score, tried, key = hill_climb(key, cipher_idxs, positions, quadgrams)
        total_keys += tried

        if score > best_score:
            best_score = score
            best_key = key.copy()
            best_hit = 1
        elif score == best_score:
            best_hit += 1
            if best_hit >= consolidate:
                break

        # early stop por flag
        if STOP_ON_FLAG:
            pt = decode_with_key(ciphertext, best_key, amap, alphabet)
            if re.search(r"picoCTF\{[^}]+\}", pt):
                break

    seconds = time.time() - start
    plaintext = decode_with_key(ciphertext, best_key, amap, alphabet)

    human_fitness = (best_score / max(1, (len(cipher_idxs) - 3))) / 10.0

    return {
        "alphabet": alphabet,
        "key_plain_to_cipher": "".join(alphabet[c] for c in best_key),
        "plaintext": plaintext,
        "fitness": human_fitness,
        "rounds": rounds_done,
        "keys_tried": total_keys,
        "keys_per_second": (total_keys / seconds) if seconds > 0 else float("inf"),
        "seconds": seconds,
    }

if __name__ == "__main__":
    res = break_substitution(
        CIPHERTEXT,
        QUADGRAM_JSON_PATH,
        max_rounds=MAX_ROUNDS,
        consolidate=CONSOLIDATE,
        seed=SEED,
    )

    print("\n=== RESULT ===")
    print("Key (plain->cipher):", res["key_plain_to_cipher"])
    print("Fitness:", res["fitness"])
    print("Rounds:", res["rounds"])
    print("Keys tried:", res["keys_tried"])
    print("Keys/s:", round(res["keys_per_second"], 2))
    print("Seconds:", round(res["seconds"], 3))
    print("\n--- PLAINTEXT ---\n")
    print(res["plaintext"])

    m = re.search(r"picoCTF\{[^}]+\}", res["plaintext"])
    if m:
        print("\nFLAG FOUND:", m.group(0))
```
